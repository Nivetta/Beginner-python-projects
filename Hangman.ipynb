```python
"""PROJECT HANGMAN"""
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
from random import choice
words=['sun','star','space','nova']
comp_word=choice(words)
print(comp_word)

word_len = len(comp_word)
user_guess = ["_"] * word_len
print("Lets play Hangman")
print(user_guess)
print("Start guessing")

end_game = False
lives=6

while not end_game:
    if "_" not in user_guess:
        print("You won. Game over")
        break
    if lives>0:
        letter=input("Enter a letter : ")
        if letter.lower() in comp_word:
            user_guess[comp_word.index(letter.lower())]=letter
            print(user_guess)
        else:
            lives-=1
            print("OOPS, You just lost a life")
            print(stages[lives])
    else:
        print("Sory, you're dead")
        break
```
