### 22. Find Unique Value From Genre

```python
data.columns
```

```python
# creating a 2d list
list1 = []
for value in data["Genre"]:
    list1.append(value.split(','))
```

```python
# converting into 1d list
one_d = []
for item in list1:
    for item1 in item:
        one_d.append(item1)
one_d
```

```python
 uni_list = []
 for x in one_d:
    if x not in uni_list:
        uni_list.append(x)
uni_list
```
